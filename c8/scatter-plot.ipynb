```python
# problem : display a graph plotting time spent on a website 
# against money spent to see whether we can see a pattern.
```

```python
# 1. import data

import csv
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter 
```

```python
# 2. Prepare the data

with open('scatter.csv') as fp:
    reader = csv.reader(fp)
    data = list(reader)
```

```python
# 3. Prepare the data for plotting and plot it 

data_x = [float(x) for x, y in data]
data_y = [float(y) for x, y in data]


def format_minutes(value, pos):
    return '{}m'.format(int(value))
def format_value(value, pos):
    return '${}'.format(value)

plt.gca().xaxis.set_major_formatter(FuncFormatter(format_minutes))

plt.xlabel('Time in website')

plt.gca().yaxis.set_major_formatter(FuncFormatter(format_value))

plt.ylabel('Total Spending')

plt.scatter(data_x, data_y)
```

#### More

A scatter plot can display not only points in two dimensions, but also add a third
(area) and even a fourth dimension (color).

```python
import random
```

```python
NUM_POINTS = 100
COLOR_SCALE = ['#FF0000', '#FFFF00', '#FFFF00', '#7FFF00', '#00FF00']

data_x = [random.random() for _ in range(NUM_POINTS)]
data_y = [random.random() for _ in range(NUM_POINTS)]

size = [(50 * random.random()) ** 2 for _ in range(NUM_POINTS)]
color = [(random.choice(COLOR_SCALE)) for _ in range(NUM_POINTS)]

plt.scatter(data_x, data_y, size, color, alpha=0.5)
```
