```python
# Import module
import matplotlib.pyplot as plt 
```

```python
# 2. Create data

DATA = (
('Q1 2017', 100, 0),
('Q2 2017', 105, 15),
('Q3 2017', 125, 40),
('Q4 2017', 115, 80),
)
```

```python
# 3. Convert data into suitable format

POS = list(range(len(DATA)))
VALUESA = [valueA for label, valueA, valueB in DATA]
VALUESB = [valueB for label, valueA, valueB in DATA]
LABELS = [label for label, value1, value2 in DATA]
```

```python
# 4. Create the bar plot 

plt.bar(POS, VALUESA)

plt.bar(POS, VALUESB, bottom=VALUESA)

plt.ylabel('Sales')

plt.xticks(POS, LABELS)
```

## More

Representing stacked bars as percentages, making the bars change relative to each other, can be done
in the following way :

```python
VALUESA = [ 100 * valueA / (valueA + valueB) for label, valueA, valueB in DATA]
VALUESB = [ 100 * valueB / (valueA + valueB) for label, valueA, valueB in DATA]
```

```python
plt.bar(POS, VALUESA)

plt.bar(POS, VALUESB, bottom=VALUESA)

plt.ylabel('Sales')

plt.xticks(POS, LABELS)
```
