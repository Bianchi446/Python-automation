```python
# Problem : Creating piecharts in Python
```

```python
#1. import

import matplotlib.pyplot as plt 
```

```python
# 2. CREATE DATA

DATA = (
('Common', 100),
('Premium', 75),
('Luxurious', 50),
('Extravagant', 20),
)
```

```python
# 3. Put in appropiate format

VALUES = [value for label, value in DATA]
LABELS = [label for label, value in DATA]
```

```python
# 4. Create pie chart


plt.pie(VALUES, labels=LABELS, autopct='%1.1f%%')

plt.gca().axis('equal')
```
